```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
dataset = pd.read_csv('properties.csv')
dataset
```

```python
dataset = dataset.replace('None',np.nan)
```

```python
dataset
```

```python
dataset.isna().sum()
```

```python
del dataset['Car Space']
del dataset['Condominum Name']
del dataset['Description']
del dataset['Sub-Category']
del dataset['Link']
dataset
```

```python
dataset = dataset.rename(columns={'Price(Php)':'Price', 'Building Size(M2)':'BuildingSize', 'Land Size(M2)':'LandSize'})
dataset
```

```python
dataset=dataset.dropna()
dataset
```

```python
dataset.isna().sum()
```

```python
dataset.dtypes
```

```python
dataset.Bedrooms = dataset.Bedrooms.astype(float)
dataset.Bathrooms =dataset.Bathrooms.astype(float)
dataset.LandSize =dataset.LandSize.astype(float)
dataset.BuildingSize =dataset.BuildingSize.astype(float)


dataset.dtypes
```

```python
dataset['Bedrooms'] = dataset['Bedrooms'].apply(np.ceil) 
```

```python
dataset.Bedrooms = dataset.Bedrooms.astype(int)
dataset.Bathrooms =dataset.Bathrooms.astype(int)
dataset.LandSize =dataset.LandSize.astype(int)
dataset.BuildingSize =dataset.BuildingSize.astype(int)

dataset.dtypes
```

```python
dataset.to_csv('cleaned_data.csv')
```

```python
del dataset['Location']
```

```python
dataset = pd.get_dummies(dataset, columns = ['Category'])
dataset
```

```python
dataset.describe()
```

```python
dataset.corr()
```

```python
correlation_matrix = sns.heatmap(dataset.corr(), annot = True)

correlation_matrix.set(xlabel='Pricing Details', ylabel='Pricing Details', title = "Correlation matrix of House Pricing")

plt.show()
```

```python
x = dataset.drop(['Price'],axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(xtrain,ytrain)
```

```python
y_pred= regressor.predict(xtest)  
```

```python
print('Train Score: ', regressor.score(xtrain, ytrain))  
print('Test Score: ', regressor.score(xtest, ytest)) 
```

```python
dataset_predicted = pd.DataFrame(data=ytest)
dataset_predicted['Predicted'] = y_pred
dataset_predicted
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(r2_score(ytest,y_pred))
print(mean_squared_error(ytest,y_pred))
print(mean_absolute_error(ytest,y_pred))
```

```python
import statsmodels.api as sm
regressor_OLS=sm.OLS(ytrain,xtrain)
regressor_OLS= regressor_OLS.fit()
print(regressor_OLS.summary())
```
